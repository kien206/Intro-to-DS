## Train model

```python
# import models
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import OneHotEncoder
from sklearn.metrics import r2_score, mean_absolute_error,mean_squared_error
```

```python
from sklearn.linear_model import LinearRegression, Ridge, Lasso
from sklearn.neighbors import KNeighborsRegressor
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor, ExtraTreesRegressor
from sklearn.svm import SVR
from xgboost import XGBRegressor
from sklearn.preprocessing import StandardScaler
```

## Linear Regression

```python
step1 = ColumnTransformer(transformers=[
    ('col_tnf', OneHotEncoder(sparse=False, drop='first', handle_unknown='ignore'), [0, 1, 2, 3, 4, 5,6,7, 12,14]),
    
], remainder='passthrough')
step2 = LinearRegression()
pipe = Pipeline([('step1', step1), ('step2', step2)])
pipe.fit(X_train, y_train)
y_pred = pipe.predict(X_test)
print('R2 Score: ', r2_score(y_test, y_pred))
print('MAE Score: ', mean_absolute_error(y_test, y_pred))
print('MSE Score:', mean_squared_error(y_test,y_pred))
```

## Ridge

```python
step1 = ColumnTransformer(transformers=[
    ('col_tnf', OneHotEncoder(sparse_output=False, drop='first', handle_unknown='ignore'), [0, 1, 2, 3, 4, 5,6,7, 12,14])
], remainder='passthrough')
alpha_values=[0.01, 0.05, 0.1, 0.5, 1 ,2 , 5, 10, 50, 100, 500]
r2_scores=[]
mae_scores=[]
mse_scores=[]

for alpha in alpha_values:
    step2=Ridge(alpha=alpha)
    pipe=Pipeline([('step1',step1),('step2',step2)])
    pipe.fit(X_train, y_train)
    y_pred = pipe.predict(X_test)
    r2_scores.append(r2_score(y_test, y_pred))
    mae_scores.append(mean_absolute_error(y_test, y_pred))
    mse_scores.append(mean_squared_error(y_test,y_pred))
print("Best r2_scores: ",max(r2_scores))
print("Best MAE scores: ",min(mae_scores))
print('MSE Score:', mean_squared_error(y_test,y_pred))
#Visualize the scores:
plt.figure(figsize=(10,6))
plt.plot(alpha_values, r2_scores, 'go-', label='R2_score', linewidth=2, markersize=8)
plt.plot(alpha_values, mae_scores, 'bo-', label='MAE score', linewidth=2, markersize=8)
plt.plot(alpha_values, mse_scores, 'ro-', label='MSE score', linewidth=2, markersize=8)
plt.xlabel('Alpha')
plt.ylabel('Score')
plt.grid(True)
plt.legend()
plt.title("Visualize the score with alpha changing")
plt.show()
```

## Gradient Boost

```python
from sklearn.ensemble import GradientBoostingRegressor


step1 = ColumnTransformer(transformers=[
    ('col_tnf', OneHotEncoder(sparse_output=False, drop='first', handle_unknown='ignore'), [0, 1, 2, 3, 4, 5,6,7, 12,14])
], remainder='passthrough')
n_estimators = [1, 5, 10, 50, 100]
r2_scores = []
mae_scores = []
mse_scores=[]
for n in n_estimators:
    step2 = GradientBoostingRegressor(n_estimators=n, random_state=42)
    pipe=Pipeline([('step1',step1),('step2',step2)])
    pipe.fit(X_train, y_train)
    y_pred = pipe.predict(X_test)
    r2 = r2_score(y_test, y_pred)
    mae = mean_absolute_error(y_test, y_pred)
    mse=mean_squared_error(y_test,y_pred)
    r2_scores.append(r2)
    mae_scores.append(mae)
    mse_scores.append(mse)
plt.figure(figsize=(10, 6))
plt.plot(n_estimators, r2_scores, label='R2 Score')
plt.plot(n_estimators, mae_scores, label='MAE')
plt.plot(n_estimators,mse_scores, label='MSE')
plt.xlabel('Number of Estimators')
plt.ylabel('Accuracy')
plt.title('Accuracy of Gradient Boosting with Different Number of Estimators')
plt.grid(True)
plt.legend()
plt.show()
```

```python
step2=GradientBoostingRegressor(n_estimators=100,random_state=42)
pipe=Pipeline([('step1',step1),('step2',step2)])
pipe.fit(X_train,y_train)
y_pred=pipe.predict(X_test)
r2score = r2_score(y_test, y_pred)
mae = mean_absolute_error(y_test, y_pred)
mse=mean_squared_error(y_test,y_pred)
print('R2 score:', r2score)
print('MAE:', mae)
print("MSE: ",mse)
```

## RANDOM FOREST


```python
step1 = ColumnTransformer(transformers=[
    ('col_tnf', OneHotEncoder(sparse_output=False, drop='first', handle_unknown='ignore'), [0, 1, 2, 3, 4, 5,6,7, 12,14])
], remainder='passthrough')
n_estimators_values = [10, 50, 100, 150, 200]
r2_scores = []
mae_scores = []
mse_scores = []

for n_estimators in n_estimators_values:
    step2 = RandomForestRegressor(n_estimators=n_estimators, bootstrap=True, random_state=42, max_samples=0.5, max_features=0.75, max_depth=10)
    pipe = Pipeline([('step1', step1), ('step2', step2)])
    pipe.fit(X_train, y_train)
    y_pred = pipe.predict(X_test)
    r2score = r2_score(y_test, y_pred)
    mae = mean_absolute_error(y_test, y_pred)
    mse = mean_squared_error(y_test, y_pred)
    r2_scores.append(r2score)
    mae_scores.append(mae)
    mse_scores.append(mse)

plt.figure(figsize=(10, 6))

plt.subplot(2, 2, 1)
plt.plot(n_estimators_values, r2_scores, marker='o')
plt.title('R2 Score vs n_estimators')
plt.xlabel('n_estimators')
plt.ylabel('R2 Score')

plt.subplot(2, 2, 2)
plt.plot(n_estimators_values, mae_scores, marker='o')
plt.title('MAE vs n_estimators')
plt.xlabel('n_estimators')
plt.ylabel('MAE')

plt.subplot(2, 2, 3)
plt.plot(n_estimators_values, mse_scores, marker='o')
plt.title('MSE vs n_estimators')
plt.xlabel('n_estimators')
plt.ylabel('MSE')

plt.tight_layout()
plt.show()
```

## XG Boost

## Voting Regressor

```python
from sklearn.ensemble import VotingRegressor,StackingRegressor
step1 = ColumnTransformer(transformers=[
    ('col_tnf', OneHotEncoder(sparse_output=False, drop='first', handle_unknown='ignore'), [0, 1, 2, 3, 4, 5,6,7, 12,14]),
    
], remainder='passthrough')
rf = RandomForestRegressor(n_estimators=10, bootstrap=True, random_state=42, max_samples=0.5, max_features=0.75, max_depth=10)
gbdt = GradientBoostingRegressor(n_estimators=100, random_state=42, max_features=0.5)
xgb = XGBRegressor(n_estimators=500,random_state=42,learning_rate=0.1, max_depth=5)

step2 = VotingRegressor([('rf', rf), ('gbdt', gbdt), ('xgb',xgb)],weights=[1,1,1])

pipe = Pipeline([
    ('step1',step1),
    ('step2',step2)
])

pipe.fit(X_train,y_train)

y_pred = pipe.predict(X_test)

r2score = r2_score(y_test, y_pred)
mae = mean_absolute_error(y_test, y_pred)
mse=mean_squared_error(y_test,y_pred)
print('R2 score:', r2score)
print('MAE:', mae)
print("MSE: ",mse)
```

```python
rf = RandomForestRegressor(n_estimators=10, bootstrap=True, random_state=42, max_samples=0.5, max_features=0.75, max_depth=10)
gbdt = GradientBoostingRegressor(n_estimators=100, random_state=42, max_features=0.5)
xgb = XGBRegressor(n_estimators=500,random_state=42,learning_rate=0.1, max_depth=5)

rfpipe=Pipeline([('step1',step1),('step2',rf)])
gbrpipe=Pipeline([('step1',step1),('step2',gbr)])
xgbpipe=Pipeline([('step1',step1),('step2',xgb)])
step2 = VotingRegressor([('rf', rf), ('gbr', gbr), ('xgb', xgb)])
pipe=Pipeline([('step1',step1),('step2',step2)])
pipe.fit(X_train, y_train)
rfpipe.fit(X_train,y_train)
gbrpipe.fit(X_train,y_train)
xgbpipe.fit(X_train,y_train)
y_pred_rf = rfpipe.predict(X_test)
y_pred_gbr = gbrpipe.predict(X_test)
y_pred_xgb = xgbpipe.predict(X_test)
r2score_rf = r2_score(y_test, y_pred_rf)
r2score_gbr = r2_score(y_test, y_pred_gbr)
r2score_xgb = r2_score(y_test, y_pred_xgb)

fig, (ax1, ax2,ax3) = plt.subplots(3, 1, figsize=(10, 15))
regressors_r2 = ['Random Forest', 'Gradient Boosting', 'XGBoost']
r2_scores = [r2score_rf, r2score_gbr, r2score_xgb]

ax1.bar(regressors_r2, r2_scores)
ax1.set_xlabel('Regressor')
ax1.set_ylabel('R2 Score')
ax1.set_title('R2 Score Comparison for Each Regressor')
for i, v in enumerate(r2_scores):
    ax1.text(i, v, str(round(v, 4)), ha='center', va='bottom')
    

mae_rf = mean_absolute_error(y_test, y_pred_rf)
mae_gbr = mean_absolute_error(y_test, y_pred_gbr)
mae_xgb = mean_absolute_error(y_test, y_pred_xgb)

mse_rf = mean_squared_error(y_test, y_pred_rf)
mse_gbr = mean_squared_error(y_test, y_pred_gbr)
mse_xgb = mean_squared_error(y_test, y_pred_xgb)
regressors_mae = ['Random Forest', 'Gradient Boosting', 'XGBoost']
mae_scores = [mae_rf, mae_gbr, mae_xgb]
mse_scores=[mse_rf,mse_gbr,mse_xgb]
ax2.bar(regressors_mae, mae_scores)
ax2.set_xlabel('Regressor')
ax2.set_ylabel('MAE Score')
ax2.set_title('MAE Score Comparison for Each Regressor')
for i, v in enumerate(mae_scores):
    ax2.text(i, v, str(round(v, 4)), ha='center', va='bottom')
regressors_mse=['Random Forest', 'Gradient Boosting', 'XGBoost']
ax3.bar(regressors_mse, mse_scores)
ax3.set_xlabel('Regressor')
ax3.set_ylabel('MSE Score')
ax3.set_title('MSE Score Comparison for Each Regressor')
for i, v in enumerate(mse_scores):
    ax3.text(i, v, str(round(v, 4)), ha='center', va='bottom')
    
fig.show()
```
